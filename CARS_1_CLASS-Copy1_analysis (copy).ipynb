```python
cat_columns=[ 'Fuel_Type', 'Transmission',
       'Owner_Type','Seats', 
        'Mileage_Unit', 'Power_Unit',
       'Engine_Value', 'Engine_Unit']
```

```python
pos=1
plt.figure(figsize=(8,8))
for i in df.columns:
    if i in cat_columns:
        x=df[i].value_counts().index
        y=df[i].value_counts().values
        plt.subplot(4,2,pos)
        plt.bar(x,y)
   
        plt.title(f"{i}")
        pos+=1
plt.suptitle('Total no of cars by different fields')
plt.tight_layout()
plt.show()
```
